```python
# 1. Create a list named shopping_list containing at least 5 initial items (e.g., "milk", "bread", "eggs", "apples", "chicken").

shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]

print(shopping_list)

len(shopping_list)
```

```python
# 2. Print the entire list, and also print the third item in the list.
 
print('The entire list is that: ',shopping_list, '\nThe third item in the list is: ', shopping_list[3])
```

```python
# 3. Add two new items to the list using an appropriate list method.



shopping_list.append('Banana')

## if I add both, rise error, and otherwise create a nested list, insert method also cause error.

shopping_list.append('Oranges')
```

```python
# 4. Remove one item from the list.

shopping_list.remove(shopping_list[0])

print('New list of items: ', shopping_list)
```

```python
# 5. Create a dictionary named item_prices. 
# The keys should be items from your shopping list (at least 4), and the values should be their prices (as floats).


item_prices = {'chicken': 550.5, 'Oranges': 330.5, 'apples': 450.5, 'Banana':200.53}


print('Items Prices dictionary :', '\n', item_prices)

# value = item_prices.values()

# print('The values from dictionary.',value)

# key = item_prices.keys()

# print('The keys from dictionary.',key)
```

```python
# 6. Print the price of one specific item from the dictionary.

print('Price of the item is: ',item_prices['chicken'])
```

```python
# 7. Print a formatted summary of your final shopping list and the price dictionary.

for keys, values in item_prices.items():
    print (f'\nFinal shopping list:\n\nKey:' , keys,  '\nValue:' ,   values)

print('\nFinal shopping list:', '\n',key , '\n\nThe items price :' ,'\n' ,value)
```
