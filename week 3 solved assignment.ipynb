<!-- 5. Missing data:
   - df.isnull()

   - Drop rows with missing Score -->


```python
df.isnull().sum()
```

```python
df['Age'].mean()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())

print(df['Age'])
```

```python
df['Score']
```

```python
df = df.dropna(subset=['Score'])

df['Score'].isnull().sum()
```

6. Selection:
