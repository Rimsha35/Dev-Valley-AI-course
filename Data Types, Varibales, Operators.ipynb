4- Write a program to:
 -Add two numbers
 -Subtract two numbers
 -Multiply two numbers
 -Divide two numbers

 

```python
user_input1 = int(input('Enter 1st number.'))

user_input2 = int(input('Enter 2nd number.'))



add_result = user_input1 + user_input2

sub_result = user_input1 - user_input2

multi_result = user_input1 * user_input2

divide_result = user_input1 /  user_input2
```

```python
print('Adding two numbers: ', add_result)

print('\nSubtracting two numbers: ', sub_result)

print('\nMultiplying two numbers: ', multi_result)

print('\nDividing two numbers: ', divide_result)
```

```python
#  5- Create a program that compares two numbers 
#  - using operators:> greater than< less than== equal to
```

```python
user_input1 = int(input('Enter 1st number.'))

user_input2 = int(input('Enter 2nd number.'))



if user_input1 > user_input2:
    print('First number is greater than 2nd. ')

if user_input1<user_input2:
    print('First number is less than 2nd.' )

if user_input1 == user_input2:

    print('Both numbers are equal.')
```
