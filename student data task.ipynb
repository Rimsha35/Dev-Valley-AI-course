```python
missing_values = ['CGPA','Attendance_Rate' ]

for col in missing_values:
    student_data[col] = student_data[col].fillna(student_data[col].mod(0))

    student_data['CGPA'] = student_data['CGPA'].fillna(student_data['CGPA'].mod(0))
    student_data['Attendance_Rate'] = student_data['Attendance_Rate'].fillna(student_data['Attendance_Rate'].mod(0))

    display(student_data.isnull().sum())
```
