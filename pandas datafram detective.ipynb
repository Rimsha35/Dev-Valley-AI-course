```python
#1. Import pandas as pd and numpy as np. 

import pandas as pd
import numpy as np
```

```python
#2. Create dictionary:


data = {
        'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
        'Age': [25,30,np.nan,35,28,np.nan],
        'Score': [85,92,78,np.nan,88,95],
        'Grade': ['A','A','B','C','B','A']
        } 

```

```python
#3. Create DataFrame df.

df = pd.DataFrame(data)
data
```

```python
#4. Perform:   - df.head(3)   - df.shape   - df.columns   - df.dtypes

print('First three rows:', df.head(3))
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
#5. Missing data:   - df.isnull()   - Fill Age with mean   - Drop rows with missing Score 


df.isnull().sum()
```

```python
mean_value =df['Age'].mean()

df['Age'].fillna(mean_value, inplace=True)
```

```python
df.dropna(subset=['Score'])
```

```python
#6. Selection:   - Select Name and Score   - Select rows where Score > 85   - Set Name as index 


select_result = df[['Name', 'Score']][df['Score']>85].set_index('Name')

select_result
```

```python
#7. Operations:   - Unique grades   - Grade counts   - Sort by Score descending
df['Grade'].unique()
```

```python
df['Grade'].value_counts()
```

```python
df['Score'].sort_values(ascending=False)
```
